결국 각 데이터의 분산(흩어진 정도)을 두 변수 간의 공분산(함께 흩어진 정도)로 얼마나 설명할 수 있는가? 를 의미한다. 만일 r이 1이라면 x와 y를 묶어서 봤을 때 x와 y 각각에 대한 정보를 전부 전달할 수 있다는 것이다.

#정규식 기초

정규식(Regular Expression, Regex): 특정한 규칙을 가진 문자열의 집합을 표현하기 위해 쓰이는 형식언어

정규표현식 기초 문법





---

 . :  \n을 제외한 모든 문자와 일치

---

[  : 문자클래스 시작

---
]  : 문자클래스 종료



---

(  : 문자 그룹 시작

---

)  : 문자 그룹 종료

---

\*, ?  : 문자 그룹이나 바로 앞 문자가 0번 이상 반복됨. 즉 있거나 없거나  +  : 문자 그룹이나 바로 앞 문자가 1번 이상 반복됨. 즉 최소한 하나 있는 상황.

---

^  : 문자열의 시작과 일치


---

$  : 문자열의 끝과 일치

---

\-  : 문자열의 범위

---




간단한 예시 몇 가지


1. .^ : 문자열의 시작이 엔터를 제외한 모든 문자. 즉 빈 문자열 제외하고는 전부 맞음.

2. a+ : a(바로 앞 문자)가 1번 이상 반복됨.

3. [a-z] : 모든 소문자 알파벳을 의미

4. a.b : a와 b 사이에 아무 문자나 가능. aab, axb 등이 해당.

5. ca*t : c와 t 사이에 a가 0번 이상 반복됨. ct, cat, caaaat 등이 해당.

6. [a-z][0-9] : 소문자 알파벳 한 글자 다음 일의 자리 정수 한 글자가 반복되는 것. x3, t6 등이 해당.


```python
import re #정규표현식을 파이썬에서 활용할 수 있는 모듈
#알파벳과 숫자로 혼합된 문자열에서 영어만 추출해 보자.

rule = "[a-zA-Z]" #소문자 알파벳 또는 대문자 알파벳.  

p = re.compile(rule) #정규식을 컴파일해 객체로 만든다.

test_str = "Infected4098" 

p.findall(test_str) #findall 함수는 정규식과 매치되는 모든 문자를 리스트 형태로 불러옴.
```

```python
#특정 문자열이 매치되는 케이스가 있는지만 찾아보려면 match를 쓴다.
#사용자 아이디에 한글이 포함되어 있는지 확인해 보자.

true_str = "이용준123"
false_str = "Infected4098"

rule = "[ㄱ-ㅎ가-힣ㅏ-ㅣ]" #자음, 모음 등 한글 모두 처리할 수 있는 문자 객체 만들기.
p = re.compile(rule) 
print(p.match(true_str), "\n")
print(p.match(false_str))
```

```python
#위 이미지 속 요청 프로토콜에서 볼 수 있듯, authkey에 인증키를 string형태로 입력하고, searchdate는 검색을 요청하고자 하는 날짜, data는 string 형태로 원하는 데이터를 불러올 수 있게 되어 있다.
#2018년 1월 2일의 환율에 대해 알아보자. 한 시점의 정보를 알아내려는 것이므로 정적 크롤링.

key_test = "hKCReZdKTryJnjFyqrb5mGhpmc4VRDjX" #활용신청(request)를 해 발급받은 인증 키
searchdate_test = "20180102"
data_test = "AP01"

request_url = "https://www.koreaexim.go.kr/site/program/financial/exchangeJSON?authkey="+str(key_test)+"&searchdate="+str(searchdate_test)+"&data="+str(data_test)

request_url
```

```python
import numpy as np
import pandas as pd
import requests #정적 크롤링에서 정보를 요청할 수 있게 해주는 모듈.
from bs4 import BeautifulSoup as bs #제공(response)받은 정보를 마크업 언어의 형태로 변환해 파싱할 수 있게 해주는 모듈.
import json #마크업 언어의 일종. 쉽고 직관적임.
"""
Json은 object, array, key-value의 세 가지로 구성된 데이터 저장 마크업 언어. 
object는 {}. 즉 중괄호로 묶인 정보를 의미.
array는 [], 즉 대괄호로 묶인 정보를 의미.
key-value는 딕셔너리의 키와 값과 같은 형태의 데이터를 의미.


"""
response = requests.get(request_url) #url을 입력한 뒤 requests.get(url)을 해 정보를 제공받음.

text = response.text #제공받은 데이터를 text로 변환.

json_text = json.loads(text) #text로 변환한 데이터를 json 형태로 변환.

#아래의 json파일은 array는 없지만 여러 object와 key-value값이 혼재한 json형식이다.
json_text
```
