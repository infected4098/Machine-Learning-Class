# Linear Regression pt.1 ( Linearity and models)

(linear map),      :      (Additivity) (Homogeneity)       .

   $$ Additivity \\ f(x+y) = f(x) + f(y) $$

$$ Homogeneity \\ \alpha\, f(x) = f(\alpha\, x)$$

   $$ \hat{y} = b_{1}x+b_{0}$$

   
$$ \hat{y} = b_{1}x_{1} + b_{2}x_{2} + ... + b_{n}x_{n}+b_{0} $$

     
$$ \hat{y} = \bar{b}\bar{x}+b_{0} $$

  (, , ... )

#Linear Regression pt.2 (Loss function and optimizations)

```python
#   
#  ?   . 
```

$$ Ordinary \ Least \ Squares \ Loss = \sum{(y - \hat{y})^2} $$

```python
#
#argmin          . 
```

$$ \underset{b_{0}, b_{1}, ... , b_{n}}{\operatorname{argmin}} 
(\sum{(y-(b_{1}x_{1} + b_{2}x_{2} + ... + b_{n}x_{n}+b_{0}))^2}) $$

```python
#     . 
from sklearn.linear_model import LinearRegression #
from sklearn.model_selection import train_test_split # ,   
from sklearn.preprocessing import PolynomialFeatures #
from sklearn.metrics import r2_score,mean_squared_error #. Loss .
from sklearn.ensemble import RandomForestRegressor #Randomforest 
```

```python
lr = LinearRegression(fit_intercept = True) # class. fit_intercept = True  b0 . 
lr.fit(X_train, y_train) # train data 

```
